### Cleaning the countries and i94addr mapping for Immigration data

```python
import numpy as np
import pandas as pd
pd.set_option("max_rows", 500)
```

```python
cit_res_codes = []
country = []
continent = []
```

```python
with open('../I94_Countries.txt') as f:
    for line in f:
        temp = line.strip('\n').split("=")
        cit_res_codes.append(temp[0].strip().strip("'"))
        exp = temp[1].strip().strip("'").lower()
        continent.append(temp[2].strip().strip("'"))
        country.append(exp)
```

```python
df_citres = pd.DataFrame({'cit_res_code':cit_res_codes, 
                           'country': country,
                         'continent':continent})
```

```python
df_citres
```

```python
df_citres.to_csv('../Cleaned Data/Country_Codes.csv', index=False)
```

#### I94addr

```python
i94addr = []
i94addr_US_state = []
```

```python
with open('../US_States.txt') as f:
    for line in f:
        temp = line.strip('\n').split("=")
        i94addr.append(temp[0].strip().strip("'"))
        exp = " ".join([elem.capitalize() for elem in temp[1].strip().strip("'").split(" ")])
        i94addr_US_state.append(exp)
```

```python
df_i94addr = pd.DataFrame({'i94addr':i94addr, 
                           'i94addr_US_state': i94addr_US_state})
```

```python
df_i94addr
```

```python
df_i94addr.to_csv("I94_addr.csv", index=False)
```
